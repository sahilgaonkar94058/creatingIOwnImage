```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
myimg = np.zeros((512,512,3), dtype=np.uint8)
plt.imshow(myimg)
```

```python

start_point= (0, 0)
end_point = (500, 500)

start_point_2= (500, 0)
end_point_2 = (0, 500)

start_point_3= (250, 0)
end_point_3 = (250, 500)

start_point_4= (250, 250)
end_point_4 = (500,250)

r_color = (255, 255, 0)

color = (0, 255, 0)

thickness = 9

image =cv2.line(myimg, start_point, end_point, color,thickness)
image =cv2.line(myimg, start_point_2, end_point_2, color,thickness)
image =cv2.line(myimg, start_point_3, end_point_3, color,thickness)
image =cv2.line(myimg, start_point_4, end_point_4, r_color,thickness)

plt.imshow(image)
```

```python

f_size = 20
s_size = 30
# s2_zize =500
while(f_size<=500 and s_size<=500):
    f_size += 20
    s_size +=20
    center_coordinates = (f_size, s_size)
    radius = 30
    color = (255, 0, 0)
    thickness = 2
    image_1 = cv2.circle(image, center_coordinates, radius, color, thickness)
    plt.imshow(image_1)
```
